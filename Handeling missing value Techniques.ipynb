# Handeling missing values Technique 2

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("train.csv")
```

```python
df.shape
```

```python
pd.set_option('display.max_column',None)
#pd.set_option('display.max_row',None)
```

```python
null_clm
```

```python
df3=df.drop(columns=null_clm)
```

```python
df3.isnull().sum()/df.shape[0]*100
```

```python
df3_num=df.select_dtypes(include=['int64','float64'])
```

```python
df3_num.head(2)
```

```python
plt.figure(figsize=[25,25])
sns.heatmap(df3.isnull())
```

```python
df3.isnull()
```

```python
df3[df3.isnull().any(axis=1)]
```

```python
pd.set_option('max_rows',None)
df3.isnull().sum()
```

```python
df3[df3.isnull().any(axis=1)]
```

```python
missing_num_vars=[var for var in df3_num.columns if df3_num[var].isnull().sum()>0]
missing_num_vars
```

```python
sns.set()
for i ,var in enumerate(missing_num_vars):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins=20,kde_kws={'linewidth':5,'color':'#DC143C'})
```

```python
df4_num_mean=df3_num.fillna(df3_num.mean())
```

```python
df4_num_mean.isnull().sum()
```

```python
for i,var in enumerate(missing_num_vars):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins=20,kde_kws={'linewidth':8,'color':'red'})
    sns.distplot(df4_num_mean[var],bins=20,kde_kws={'linewidth':5,'color':'green'})
    plt.legend()
```

```python
df4_num_med=df3_num.fillna(df3_num.median())
```

```python
for i,var in enumerate(missing_num_vars):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins=20,kde_kws={'linewidth':7,'color':'red'})
    sns.distplot(df4_num_mean[var],bins=20,kde_kws={'linewidth':5,'color':'green'})
    sns.distplot(df4_num_med[var],bins=20,kde_kws={'linewidth':3,'color':'white'})
    plt.legend()
```

```python
df_concat=pd.concat([df3_num[missing_num_vars],df4_num_mean[missing_num_vars],df4_num_med[missing_num_vars]],axis=1)
```

```python
df_concat[df_concat.isnull().any(axis=1)]
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('train.csv')
```

```python
cat_var=df.select_dtypes(include='object')
cat_var.head(5)
```

```python
miss_val_per=cat_var.isnull().mean()*100
```

```python
miss_val_per
```

```python
null_var=miss_val_per[miss_val_per>20].keys()
```

```python
null_var
```

```python
cat_var=cat_var.drop(columns=null_var)
```

```python
pd.set_option('display.max_rows',None)
```

```python
isnull_per=cat_var.isnull().mean()*100
```

```python
cat_var.shape
```

```python
miss_vars=isnull_per[isnull_per>0].keys()
```

```python
miss_vars
```

```python
cat_var['MasVnrType'].value_counts()
```

```python
cat_var['MasVnrType'].fillna(cat_var['MasVnrType'].mode()[0])
```

```python
for var in miss_vars:
    cat_var[var].fillna(cat_var[var].mode()[0],inplace=True)
    print(var,'=',cat_var[var].mode()[0])
```

```python
cat_var.isnull().sum()
```

```python
plt.figure(figsize=(16,9))
for i,var in enumerate(miss_vars):
    plt.subplot(4,3,i+1)
    plt.hist(cat_var[var],label='Imput')
    plt.hist(df[var].dropna(),label='original')
    plt.legend()
```

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
train.shape
```

```python
test.shape
```

```python
X_train=train.drop(columns='SalePrice')
y_train=train['SalePrice']
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
num_vars=X_train.select_dtypes(include=['int64','float64']).columns
```

```python
num_vars
```

```python
X_train[num_vars].isnull().sum()
```

```python
imputer_mean=SimpleImputer(strategy='mean')
```

```python
imputer_mean.fit(X_train[num_vars])
```

```python
imputer_mean.statistics_
```

```python
imputer_mean.transform(X_train[num_vars])
```
